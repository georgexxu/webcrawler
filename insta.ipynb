```python
import urllib.request
import urllib.error
import json
import pprint
import csv
from urllib.parse import quote
```

```python

# coding: utf-8

# In[44]:



# In[34]:

class search:
    # 初始化方法，定义一些变量
    def __init__(self):      
        
        self.user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:57.0) Gecko/20100101 Firefox/57.0'
        
        # 初始化headers
        self.headers = {'User-Agent': self.user_agent}
        # 存放程序是否继续运行的变量
        self.enable = False
        
    def processNames(self,old):
        return old


    # 传入某一页的索引获得页面代码
    def getpage(self,name):
        name=quote(name)

        url = 'https://www.instagram.com/web/search/topsearch/?context=blended&query={}'.format(name)
            # 构建请求的request
        #print(url)
        request = urllib.request.Request(url, headers=self.headers)
            # 利用urlopen获取页面代码
        response = urllib.request.urlopen(request)
            # 将页面转化为UTF-8编码
        pageCode = response.read().decode('utf-8')
        return pageCode
    
    def getNumPost(self,name):
        contents=json.loads(self.getpage(name))
        mediaCount=contents['hashtags'][0]['hashtag']['media_count']
        return mediaCount
   
    def loadRestaurant(self):
        with open('ChineseRestaurant.csv',encoding="utf-8-sig") as csv_file:
            Restreader = csv.reader(csv_file)
            header = next(Restreader)
            RestList=list(Restreader)
        return RestList

    def longList(self):
        RestList=self.loadRestaurant()
        i=1;
        for RestInfo in RestList:   
            i+=1
            if (i>1000):
                break
            try:
                RestInfo.append(str(self.getNumPost(RestInfo[0])))
                print('{}:{}'.format(RestInfo[0],RestInfo[4]))
            except Exception as e:
                RestInfo.append('0')
                print('{}:{}'.format(RestInfo[0],RestInfo[4]))
                continue
        return RestList
#        return longList_
    
    
pages= search() 
#print(pages.loadRestaurant())
PostCountChineseRestaurant=pages.longList()


```

```python
with open('ChineseRestaurant.csv','w',newline='',encoding="utf-8-sig") as csv_file:
    Writer = csv.writer(csv_file)
    Writer.writerow(['name','priceUI','reviewCount','shortenUrl','instagramCount'])

    for row in PostCountChineseRestaurant:
        print(row)
        Writer.writerow([row[0],row[1],row[2],row[3],row[4]])
```
